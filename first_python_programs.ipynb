```python
#a program that converts days into seconds

#Input the number of days
days = int(input("input the number of days:"))

#calculate the days into seconds
seconds =  days*60*60

#print the result
print(days, "days is equal to", seconds, "seconds.")
```

```python
#a program that calculate the volume of a sphere

#Input the radius
radius = float(input("Enter the radius of the sphere:"))

#calculate the volume
pi = 3.14159
volume = (4/3) * pi * (radius ** 3)

#print the result
print(f"Volume of the sphere with radius {radius} is {volume}")
```

```python
#function to calculate the area of a square
def area(side):
    return side * side

#function to calculate the perimeter of a square
def perimeter(side):
    return 4 * side

#main program
#ask user to input the length
side_length = float(input("Enter the side length of the square:"))

#calculate the perimeter of the square and rectangle
square_area = area(side_length)
square_perimeter = perimeter(side_length)

#print the result
print(f"The area of the square is {square_area}")
print(f"The perimeter of the square is {square_perimeter}")
```

```python
#a function that checks if a character input by a user is uppercase or lowercase
def check_character(character):
  if character.isupper():
    return "uppercase"
  elif character.islower():
      return "lowercase"

#main program
letter = input("please input a letter:")

#the ouput
result = check_character(letter)
print(f"The letter '{letter}' is {result}.")
```

```python
# Create an empty list to store values
values = []

#ask user for 5 values
for count in range(5):
    number = float(input(f"Enter value {count}: "))
    values.append(number)

#add up all the numbers
total = sum(values)

# Calculate the average
average = total / len(values)

# Display the result
print("Values entered are:", values)
print("The average is:", average)
```
